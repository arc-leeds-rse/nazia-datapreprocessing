## Nazia's stuff

* sum of every row, saved as a new column 
* order by sum of rows from low to high
* remove every row where the sum equals 0 (save this as a new dataframe)
* check that column 0 has no duplicate names, remove all duplicates
 
* split the dataframe into 3 separate files based on column names. 

* All column names which contain .TP. are primary tumours, .R1. are Recurrent 1 tumours, .R2.or .R3. are all other recurrent.
* 3 separate files: 
* primary tumours
* recurrent 1 tumours
* all other recurrent 

```python
import pandas as pd
```

```python
# import tab delimited file
filename = 'gene_tpm_matrix_all_samples.tsv'
df = pd.read_csv (filename, delimiter='\t')
```

```python
# df
```

```python
# Create new column called total which is sum across the rows
df['total'] = df.sum(axis=1)
```

```python
# df['total']
```

```python
# Sort inplace by total field
df.sort_values("total", axis = 0, ascending = True, inplace = True, na_position ='last')

df
```

```python
# Drop rows with zero in total column
indexTotal = df[df['total'] == 0.0].index
indexTotal

# Check length of df
len(df)
```

```python
# Drop zero columns
df_dropped = df.drop(indexTotal)
```

```python
len(df)
```

```python
len(df_dropped)
```

```python
# Remove duplicates in Gene_symbol
df_dropped.sort_values("Gene_symbol", inplace = True)
len(df_dropped)
```

```python
# df_dropped
df_dropped.drop_duplicates(subset='Gene_symbol', keep='first')
len(df_dropped)
# No duplicates in Gene_symbol?
```

```python
# Now make the Gene_symbol the index column
df_dropped.set_index('Gene_symbol', inplace=True)
df_dropped
```

```python
primary_tumours = df_dropped.filter(regex='\.TP\.')
```
